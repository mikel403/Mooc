```python
def validar_dni(dni):
    letras = "TRWAGMYFPDXBNJZSQVHLCKE"
    
    if pd.isna(dni):
        return True  # Aceptamos NaN
    
    dni = str(dni).strip().upper()
    
    # 1. Verificar que tenga 9 caracteres
    if len(dni) != 9:
        return False
    
    # 2. Verificar que los primeros 8 son dígitos
    numero = dni[:8]
    if not numero.isdigit():
        return False
    
    # 3. Verificar que el último carácter es una letra
    letra = dni[8]
    if not letra.isalpha():
        return False

    # 4. Comprobar si la letra es la correcta
    letra_correcta = letras[int(numero) % 23]
    return letra == letra_correcta

sucio_filtrado = sucio[sucio["Número de DNI"].apply(validar_dni)]
# Crear lista de los 8 primeros caracteres de los DNIs válidos en limpio
dni_8_limpio = [str(dni).strip()[:8] for dni in limpio["Número de DNI"] if pd.notna(dni)]

# Crear un nuevo DataFrame con los que no estén en la lista
df_diferentes = sucio_filtrado[
    sucio_filtrado["Número de DNI"]
    .apply(lambda x: str(x).strip()[:8] not in dni_8_limpio if pd.notna(x) else False)
]
df_diferentes
```

```python
import ast

def limpiar_lista_de_cursos(cadena, separador=";"):
    """
    Convierte string separado por `;` en lista limpia de cursos.
    """
    if pd.isna(cadena):
        return []
    return [curso.strip() for curso in str(cadena).split(separador) if curso.strip()]

def convertir_str_a_lista(cadena):
    """
    Convierte string con formato lista a lista real. Ej: "['Curso 1', 'Curso 2']"
    """
    if pd.isna(cadena):
        return []
    try:
        return ast.literal_eval(cadena)
    except Exception:
        return []
    
def transformar_cursos(cursos, mapeo_nombres=mapeo_nombres, nombres_nuevos_validos=nombres_nuevos_validos):
    cursos_actualizados = []
    cursos_no_validos = set()

    for curso in cursos:
        curso = limpiar_texto(curso)
        if curso in mapeo_nombres:
            cursos_actualizados.append(mapeo_nombres[curso])
        elif curso in nombres_nuevos_validos:
            cursos_actualizados.append(curso)
        elif "Farmacoterapia en paciente crítico" in curso:
            cursos_actualizados.append("Farmacoterapia en paciente crítico (SEFH)")
        elif "Lectura Crítica de la Evidencia Clínica" in curso and len(curso) == 41:
            cursos_actualizados.append("Lectura Crítica de la Evidencia Clínica *")
        else:
            cursos_no_validos.add(curso)

    return set(cursos_actualizados), cursos_no_validos

def comparar_cursos_transformados(df1, df2, col1="cursos", col2="cursos"):
    diferencias = []
    cursos_no_validos_global = set()

    # Asegúrate de que el índice sea ID y único
    df1 = df1[~df1.index.duplicated(keep="first")]
    df2 = df2[~df2.index.duplicated(keep="first")]

    for idx in df1.index.intersection(df2.index):
        cursos_1_raw = convertir_str_a_lista(df1.loc[idx, col1])
        cursos_2_raw = limpiar_lista_de_cursos(df2.loc[idx, col2])

        cursos_1, no_validos_1 = transformar_cursos(cursos_1_raw)
        cursos_2, no_validos_2 = transformar_cursos(cursos_2_raw)

        cursos_no_validos_global.update(no_validos_1)
        cursos_no_validos_global.update(no_validos_2)

        if cursos_1 != cursos_2:
            diferencias.append({
                "index": idx,
                "df1_cursos": cursos_1,
                "df2_cursos": cursos_2,
                "solo_en_df1": cursos_1 - cursos_2,
                "solo_en_df2": cursos_2 - cursos_1,
            })

    df_diferencias = pd.DataFrame(diferencias)

    if cursos_no_validos_global:
        print("❗ Cursos no válidos encontrados:")
        for curso in sorted(cursos_no_validos_global):
            print("-", curso)

    return df_diferencias

df_diferencias = comparar_cursos_transformados(limpio, df_iguales, col1="cursos", col2="cursos")

if df_diferencias.empty:
    print("✅ Todos los cursos coinciden tras la transformación.")
else:
    print("❌ Hay diferencias:")
    print(df_diferencias[["index", "solo_en_df1", "solo_en_df2"]])
```

```python
def contiene_curso(row, curso):
    try:
        cursos_row = json.loads(row["cursos"])
        return curso in cursos_row
    except:
        print(f"Error en fila con correo: {row.get('Correo final', 'sin correo')}")
        return False

for curso in cursos:
    print(f"Procesando curso: {curso}")
    curso_drive = pd.read_excel(f"cursos/{curso}.xlsx")
    curso_save = curso

    # Correcciones de nombre si es necesario
    if curso == "Lectura crítica de la evidencia clínica, avanzado (LCEC 2) (CASPe)":
        curso = "Lectura crítica de la evidencia clínica, avanzado (LCEC 2) * (CASPe)"
    elif curso == "Lectura crítica de la evidencia clínica, básico (LCEC 1) (CASPe)":
        curso = "Lectura crítica de la evidencia clínica, básico (LCEC 1) * (CASPe)"

    # Añadir columna tiene_curso
    limpio["tiene_curso"] = limpio.apply(lambda row: contiene_curso(row, curso), axis=1)

    # Obtener la primera fecha por persona
    fechas_primera = (
        limpio.sort_values("Hora de inicio")  # ordenar por fecha
        .groupby("Correo final")["Hora de inicio"]
        .first()
        .reset_index()
        .rename(columns={"Correo final": "correo", "Hora de inicio": "Fecha"})
    )

    # Agrupar marcados por persona
    participantes_con_curso = (
        limpio.groupby("Correo final")["tiene_curso"]
        .max()
        .reset_index()
    )

    # --- Diferencias 1: personas que marcaron el curso pero no están en Drive ---
    df_resultado = participantes_con_curso[participantes_con_curso["tiene_curso"]]
    df_diferentes = df_resultado[~df_resultado["Correo final"].isin(curso_drive["correo"])]
    df_diferentes = df_diferentes.rename(columns={"Correo final": "correo"})
    df_diferentes = df_diferentes.merge(fechas_primera, on="correo", how="left")
    df_diferentes[["correo", "Fecha"]].to_excel(
        f"diferencias/{curso_save}_faltan_en_miniExcel.xlsx", index=False
    )

    # --- Diferencias 2: personas en el Drive que no marcaron el curso ---
    correos_drive = curso_drive["correo"].unique()
    participantes_no_marcaron = participantes_con_curso[
        (participantes_con_curso["Correo final"].isin(correos_drive)) &
        (~participantes_con_curso["tiene_curso"])
    ]
    df_no_marcados = participantes_no_marcaron.rename(columns={"Correo final": "correo"})
    df_no_marcados = df_no_marcados.merge(fechas_primera, on="correo", how="left")
    df_no_marcados[["correo", "Fecha"]].to_excel(
        f"diferencias/{curso_save}_no_marcado.xlsx", index=False
    )

    print(f"- {len(df_diferentes)} personas con curso marcado pero no en miniExcel.")
    print(f"- {len(df_no_marcados)} personas en miniExcel que no marcaron el curso.")
```
